```python
import pandas as pd
df = pd.read_csv('datasets/so_survey_subset.csv')
df
```

```python
df1 = df.Age.dropna()
```

```python
df1  = list(df1)
```

```python
fig, ax = plt.subplots()

# Data for the histogram plot
ages = df1
#ax.hist(x=ages)

bins = [0, 20, 30, 40, 50, 60, 70, 80, 90, 100]
ax.hist(x=ages, bins=bins, edgecolor='black', color='purple',
        log = True, histtype = 'bar')
#ax.hist(x=ages, bins=bins, edgecolor='black', color='green', histtype='bar',orientation='vertical')
plt.title("Ages of SO Survey Respondents")

plt.xlabel("Ages")
plt.ylabel('Total Respondents')

plt.grid(True)
plt.show()
```

```python
temp = df.Age.dropna()
mylist = list(temp)
```

```python
fig, ax = plt.subplots()

# Data for the histogram plot
ages = mylist
#ax.hist(x=ages)

bins = [0, 20, 30, 40, 50, 60, 70, 80, 90, 100]
ax.hist(x=ages, bins=bins, edgecolor='black', color='purple',
        log = True, histtype = 'bar')
#ax.hist(x=ages, bins=bins, edgecolor='black', color='green', histtype='bar',orientation='vertical')
plt.title("Ages of SO Survey Respondents")

median_age = temp.median()
average_age = temp.mean()


plt.axvline(median_age, color='red', lw= 2, label='Median Age')
plt.axvline(average_age, color='yellow', linewidth=2, label='Average Age')
plt.legend()

plt.xlabel("Ages")
plt.ylabel('Total Respondents')

plt.grid(True)
plt.show()
```

```python
from PIL import Image
rgb_img = Image.open("datasets/arif1_color.jpg")
rgb_img
```

```python
import numpy as np
grey_img = rgb_img.convert('L')
grey_img_array = np.array(grey_img)
grey_img_array
```

```python

fig = plt.figure()
ax = fig.add_subplot()

bins=[0,50,100,150,200,255]
ax.hist(x=grey_img_array, bins=bins)


# add title
plt.title("Histogram of a Grey Scale Image")

plt.xlabel("Grey Levels")
plt.ylabel('Pixel Frequency')
plt.grid(True)
# show plot
plt.show()
```
